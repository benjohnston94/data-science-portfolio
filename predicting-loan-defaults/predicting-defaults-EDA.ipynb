### Categorical features

```python
# categorical variable values
def get_categorical_values(df, target_col: str = "Credit Default"):
    """
    Sum up the value counts for all categorical columns (including target)
    """
    for col in df.columns:
        if df[col].dtype == "object":
            print(col)
            print(df[col].value_counts())
            print("\n")
        elif col == target_col:
            print(col)
            print(df[col].value_counts())
            print("\n")
        else:
            pass

get_categorical_values(df)
```

### Obversations from categorical variables
* Home ownership - 'have mortgage' can probably be grouped with home mortgage
* Years of work could be made continuous
* Purpose - lost of categories; debt consolidation could probably just be made into a standalone boolean flag
* Term could be 'short/long term' flag
* Credit default is quite imbalanced

### Numeric features

```python
# distributions
def show_histograms(df):
    """
    Shows the distribution of data for each columns separately
    """
    numeric_cols = [col for col in df.columns if df[col].dtype != "object"]
    numeric_col_ct = len(numeric_cols)
    
    f, ax = plt.subplots(numeric_col_ct, 1, figsize=(20, numeric_col_ct*4))
    
    for i, col in enumerate(numeric_cols):
        sns.distplot(df[col], kde=False, ax=ax[i])
    
    plt.show()
    
    pass

# going to write it here first because need to test a bit

show_histograms(df)
```

### Obversations from distributions
* Most distributions are right-skewed
* For current loan balance, **there is a group of customers with around 100,000,000 dollars of loan balance** - these are outliers and this may be a value that's imputed in certain conditions (e.g. when data is missing). We should investigate this a bit before moving on
* Credit score values seem to be either very low or very high
* We may need to variable which could cause leakage with the target variable (e.g. credit score, number of credit problems, bankruptcies)

### Quick digression: exploring large loan balance outliers

```python
print(df["Current Loan Amount"].value_counts(), "\n") # this gives 870 customers with 99999999 loan balance

large_loan_bal = df[df["Current Loan Amount"] == 99999999].copy()

# explore dataframe
get_categorical_values(large_loan_bal)
show_histograms(large_loan_bal)
```

### Loan balance outliers: conclusion

It's hard to say exactly what these cases are. In terms of correlations, this subset of customers seem to have low credit scores and none of them have defaulted. I'd argue that this is an imputed which is assigned in certain conditions - of which there are many possible options (e.g defect customers, temporary accounts, dummy customers) - it's impossible to say. In real life, we'd check with the stakeholders or the owners of this data what this value means and how to deal with it, but that's not possible in this case. **It makes sense to exclude these cases as it's likely a data issue, and they are all 'negative target cases' which is not a representative sample.** (will remove later in data manipulation section)

### Correlations

```python
# correlations
f, ax = plt.subplots(1, 1, figsize=(15,10))

correlations = df.corr()
sns.heatmap(correlations, annot=True, fmt=".2f", ax=ax)

ax.set_title("Correlation between variables")

plt.show()
```

### Observations
* Credit score is fairly highly correlated with credit default which will likely cause leakage - we should drop this. Also, credit score likely comes from another model which may be very similar to what we're creating so this is another reason to exclude it
* Correlation between number of credit problems and bankcruptcies. This makes a lot of sense because their definitions are similar. If we use a linear model (e.g. logistic regression) it may wise to exclude one to avoid multicolinearity (it also isn't likely to significantly decrease predictive performance). 
* Correlation between number of credit problems and tax liens (same points as above)
* Monthly debt has a medium correlation with a few other variables
